```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fitness_df = pd.read_csv("athletes.csv")
fitness_df
```

```python
#creation and cleaning of deadlift by age
age_deadlift_df = pd.DataFrame(fitness_df['age']) #adding age to dataframe
age_deadlift_df = age_deadlift_df.join(fitness_df['deadlift']) # joining deadlift
age_deadlift_df = age_deadlift_df.dropna() # dropping nan values
age_deadlift_df = age_deadlift_df[age_deadlift_df['deadlift']<1200] #cleaning lift data
age_deadlift_df = age_deadlift_df[age_deadlift_df['age']<60] #cleaning age data
age_deadlift_df = age_deadlift_df[age_deadlift_df['deadlift']>50] #setting minimum deadlift

#linear regression (simple edition)
x, y = age_deadlift_df['age'].to_list(), age_deadlift_df['deadlift'].to_list()
coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 
# poly1d_fn is now a function which takes in x and returns an estimate for y

plt.plot(x,y, 'co', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker

plt.xlim(0,60)
print(age_deadlift_df['deadlift'])
age_deadlift_df.groupby('age')['deadlift'].describe() #shows 
```

```python
#creation and cleaning of 400m run between 40 and 150 seconds by age no greater than 60
age_run400_df = pd.DataFrame(fitness_df['age']) #ading age to dataframe
age_run400_df = age_run400_df.join(fitness_df['run400']) # joining 
age_run400_df = age_run400_df.dropna() # dropping nan values
age_run400_df = age_run400_df[age_run400_df['run400']<150] #cleaning maximum
age_run400_df = age_run400_df[age_run400_df['age']<690] #cleaning age data
age_run400_df = age_run400_df[age_run400_df['run400']>40] #setting minimum 

#linear regression (simple edition)
x, y = age_run400_df['age'].to_list(), age_run400_df['run400'].to_list()
coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 
# poly1d_fn is now a function which takes in x and returns an estimate for y

plt.plot(x,y, 'co', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker

plt.xlim(0,60)

age_run400_df.groupby('age')['run400'].describe()
```

```python
#creation and cleaning of 4005k run between 400 and 9000 seconds by age no greater than 60
age_run5k_df = pd.DataFrame(fitness_df['age']) #adding age to dataframe
age_run5k_df = age_run5k_df.join(fitness_df['run5k']) # joining 
age_run5k_df = age_run5k_df.dropna() # dropping nan values
age_run5k_df = age_run5k_df[age_run5k_df['run5k']<9000] #cleaning maximum
age_run5k_df = age_run5k_df[age_run5k_df['age']<60] #cleaning age data
age_run5k_df = age_run5k_df[age_run5k_df['run5k']>400] #setting minimum 

#linear regression (simple edition)
x, y = age_run5k_df['age'].to_list(), age_run5k_df['run5k'].to_list()
coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef) 
# poly1d_fn is now a function which takes in x and returns an estimate for y

plt.plot(x,y, 'co', x, poly1d_fn(x), '--k') #'--k'=black dashed line, 'yo' = yellow circle marker

plt.xlim(10,60)
plt.ylim(400,3000)

age_run5k_df.groupby('age')['run5k'].describe()
```
